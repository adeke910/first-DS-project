```python
plt.figure(figsize=(18,8))
plt.bar(x="generation", height="suicides_no",  data=master_df, color="darkslateblue")
plt.xlabel('Generation', fontdict = {'fontsize' : 18}, color="darkslateblue")
plt.ylabel('Suicides_No', fontdict = {'fontsize' : 20}, color="darkslateblue")
plt.title("Number of Suicides VS Generation", fontdict = {'fontsize' : 24}, color="darkslateblue")
plt.xticks(rotation=45, fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(18,8))
plt.scatter("year", "gdp_per_capita ($)", data= master_df, color= 'midnightblue' )
plt.title("Year VS GDP per Capita ($)", fontdict={'fontsize': 25}, color= 'midnightblue')
plt.xlabel('Year', fontdict={'fontsize': 15}, color= 'midnightblue')
plt.ylabel('GDP_per_Capita ($)', fontdict={'fontsize': 15}, color= 'midnightblue')
plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.plot("age","suicides_no","_", data=master_df, color= "navy")
plt.title("Age Variations VS Number of Suicides", fontdict={'fontsize': 25}, color= "navy")
plt.xlabel('Age', fontdict={'fontsize': 15}, color= "navy")
plt.ylabel('Suicides_No', fontdict={'fontsize': 15}, color= "navy")
plt.show()
```

```python
plt.figure(figsize=(18,8))
plt.bar("country","population", data=master_df, color="darkslateblue")
plt.xlabel('Country', fontdict = {'fontsize' : 20}, color="darkslateblue")
plt.ylabel('Population', fontdict = {'fontsize' : 20}, color="darkslateblue")
plt.title("Country VS Population", fontdict = {'fontsize' : 24}, color="darkslateblue")
plt.xticks(rotation=90, fontsize=12)
plt.yticks(fontsize=15)
plt.show()
```

```python
total=master_df.groupby("generation")["population"].agg(population="sum")
print(total)
```

```python
plt.figure(figsize=(15,8))
plt.plot('generation', 'population','_', data=master_df)
plt.title("Generation VS population", fontdict = {'fontsize' : 18})
plt.xticks(rotation=45, fontsize=15)
plt.show()
```

```python
total = master_df[["generation"]].value_counts()
total
plt.figure(figsize=(15,8))
plt.bar(x=master_df["generation"].unique(), height=total, data=master_df)
plt.title("Count per Generation", fontdict = {'fontsize' : 18})
plt.xticks(rotation=45, fontsize=15)
plt.xlabel("Generation", fontdict = {'fontsize' : 18})
plt.show()
```
